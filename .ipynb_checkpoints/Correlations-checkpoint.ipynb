```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import uproot, os, dill
import minepy as mp
```

```python
# Load Data
data_path = "../Data/data15_13TeV.00267358.physics_MinBias.30062015_v0_per_0.root"
data_file = uproot.open(data_path)
data_tree = data_file['tree']

load_variables = "v0_rxy v0_chi2 cosTheta v0_ks_mass v0_rxyErr pT ntrk_pv0 a0 calpha a0xy pv0_y".split(" ")

df = data_tree.pandas.df(load_variables, entrystop = 10**4)
```

```python
# define probability variable
model_path = "models/data_trained_xg.dat"

func_seri = dill.load(open(model_path, 'rb'))
prob_func = dill.loads(func_seri)
```

```python
df = prob_func(df)
```

```python
cols = df.columns

fig, ax = plt.subplots(figsize = (12, 12), ncols = len(cols), nrows = len(cols), sharex = 'col', sharey = 'row')

for i, col1 in enumerate(cols):
    for j, col2 in enumerate(cols):
        rang1 = np.percentile(df[col1], [1, 99])
        rang2 = np.percentile(df[col2], [1, 99])
        ax[i, j].scatter(df[col2], df[col1], alpha = 0.01)
        ax[i, j].set(xlim = rang2, ylim = rang1)

for i, col in enumerate(cols):
    ax[i, 0].set_ylabel(col)
    ax[len(cols)-1, i].set_xlabel(col)
        
fig.tight_layout()
```
